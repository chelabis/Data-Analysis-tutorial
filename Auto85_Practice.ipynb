```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import data

```python
df = pd.read_csv('imports2.csv')
df
```

## Handle NaN values

```python
df.dropna(subset=['price'], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
mean_loss = df['normalized-losses'].mean()
df['normalized-losses'].replace(np.nan, mean_loss, inplace=True)
mean_bore = df['bore'].mean()
df['bore'].replace(np.nan, mean_bore, inplace=True)
mean_stroke = df['stroke'].mean()
df['stroke'].replace(np.nan, mean_stroke, inplace=True)
mean_horsepower = df['horsepower'].mean()
df['horsepower'].replace(np.nan, mean_horsepower, inplace=True)
mean_rpm = df['peak-rpm'].mean()
df['peak-rpm'].replace(np.nan, mean_rpm, inplace=True)
df['num-of-doors'].replace(np.nan, "four", inplace=True)
df.info()
```

## Correct dtypes

```python
df = df.astype({"normalized-losses": "int64", "horsepower": "int64", "peak-rpm": "int64", "price": "int64"})
```

## Keep higher than 0.5 correlation

```python
df.drop(['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration'], axis=1, inplace=True)
df.drop(['num-of-doors', 'body-style', 'engine-location', 'height', 'engine-type'], axis=1, inplace=True)
df.drop(['stroke', 'compression-ratio', 'peak-rpm'], axis=1, inplace=True)
```

## Hot one encoding

```python
df['num-of-cylinders'].replace({"two": 2, "three": 3, "four": 4, "five": 5, "six":6, "eight": 8, "twelve": 12}, inplace=True)
df = df.join(pd.get_dummies(df['drive-wheels'], prefix='drive-wheels').astype(int)).drop('drive-wheels', axis=1)
df = df.join(pd.get_dummies(df['fuel-system'], prefix='fuel-system').astype(int)).drop('fuel-system', axis=1)
df
```

## Drop lower than 0.5 correlation

```python
df.drop(['drive-wheels_4wd', 'fuel-system_1bbl', 'fuel-system_4bbl', 'fuel-system_idi'], axis=1, inplace=True)
df.drop(['fuel-system_mfi', 'fuel-system_spdi', 'fuel-system_spfi'], axis=1, inplace=True)
df
```

```python
sns.heatmap(df.corr())
```

## Model Selection

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df.drop('price', axis=1)
y = df.price
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
reg = RandomForestRegressor(n_jobs=-1)
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
y_pred = reg.predict(X_test)
plt.scatter(y_test, y_pred, color='r')
plt.title("Prediction VS Actual Price")
plt.xlabel("Actual Car Price")
plt.ylabel("Predicted Car Price")
plt.show()
```

```python
importances = dict(zip(reg.feature_names_in_, reg.feature_importances_))
sorted_importances = sorted(importances.items(), key = lambda x: x[1], reverse=True)
sorted_importances
```
