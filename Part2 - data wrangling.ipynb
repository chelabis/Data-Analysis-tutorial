# Data Analysis tutorial: Part2 - Data Wrangling/Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('imports2.csv')
```

```python
df
```

```python
df["normalized-losses"]
```

```python
df.dropna(subset=["price"], axis=0, inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
mean = df["normalized-losses"].mean()
df["normalized-losses"] = df["normalized-losses"].replace(np.nan, mean)
```

```python
df
```

```python
mean_bore = df["bore"].mean()
df["bore"] = df["bore"].replace(np.nan, mean_bore)
mean_stroke = df["stroke"].mean()
df["stroke"] = df["stroke"].replace(np.nan, mean_stroke)
mean_horsepower = df["horsepower"].mean()
df["horsepower"] = df["horsepower"].replace(np.nan, mean_horsepower)
mean_peak = df["peak-rpm"].mean()
df["peak-rpm"] = df["peak-rpm"].replace(np.nan, mean_peak)
```

```python
df["num-of-doors"] = df["num-of-doors"].replace(np.nan, "four")
```

```python
df.info()
```

```python
df["city-mpg"] = 235/df["city-mpg"]
df.rename(columns={"city-mpg": "city-L/100km"}, inplace = True)
```

```python
df["price"] = df["price"].astype(int)
```

### Normalization

```python
df["length"] = (df["length"]-df["length"].min())/(df["length"].max()-df["length"].min())
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()
```

```python
df["height"].min()
```

### Binning

```python
bins = np.linspace(min(df["price"]), max(df["price"]), 4)
group_names = ["Low", "Medium", "High"]
df["price-binned"] = pd.cut(df["price"], bins, labels=group_names, include_lowest=True)
```
